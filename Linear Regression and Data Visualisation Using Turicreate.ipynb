## Linear Regression Using TuriCreate

```python

import turicreate as tc 
```

```python
data = tc.SFrame("insurance.csv") 
```

```python
data.head()
```

### Making a Linear Regression Model

```python
model = tc.linear_regression.create(data, target ="charges", features =['region']) 
```

```python
model.evaluate(data)
```

### Predicting Charge Using BMI

```python
bmi_person = data[data['bmi']== 27.9] 
# it will predict the charges for the person with bmi 27.9 
model.predict(bmi_person)
```

```python

# importing turicreate 
import turicreate 
  
# loading the dataset 
training_data = turicreate.SFrame.read_csv("insurance.csv")
  

training_data
```

### Visualizing Data Using TuriCreate

```python

# for data plotting 
training_data['charges'].plot(title = "Charges",  
                                  xlabel = "Amount",  
                                  ylabel = "No. of Customers") 
```

### Scatter Plot

```python
turicreate.visualization.scatter(training_data["bmi"],  
                                 training_data["charges"],  
                                 xlabel = "bmi",  
                                 ylabel = "charges")
```

```python
turicreate.visualization.scatter(training_data["children"],  
                                 training_data["charges"],  
                                 xlabel = "children",  
                                 ylabel = "charges")
```

### Heat Map

```python
turicreate.visualization.heatmap(training_data["children"],  
                                 training_data["charges"],  
                                 xlabel = "children",  
                                 ylabel = "charges") 
```

```python
turicreate.visualization.heatmap(training_data["bmi"],  
                                 training_data["charges"],  
                                 xlabel = "bmi",  
                                 ylabel = "charges") 
```

### Histogram

```python
# for histogram 
turicreate.visualization.histogram(training_data["charges"],  
                                   xlabel ="charges") 
```
